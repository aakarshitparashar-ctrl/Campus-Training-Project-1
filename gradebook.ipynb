```python
# Project 2: Student Gradebook Analyzer
import pandas as pd
import numpy as np
data = {
    "Name": ["Aman", "siya", "Rahul", "Sneha", "Karan"],
    "Assignment1": [85, 92, 76, 88, 65],
    "Assignment2": [78, 95, 80, 84, 70],
    "Assignment3": [90, 89, 70, 91, 68],
    "Assignment4": [88, 94, 75, 90, 72]
}

df = pd.DataFrame(data)

df
```

```python
# Save DataFrame to a CSV file

df.to_csv("gradebook.csv", index=False)

print("CSV file created successfully!")
```

```python
# Read the CSV file

df = pd.read_csv("gradebook.csv")

print("Data Loaded Successfully:")
df
```

```python
# Calculate student statistics

assignments = ["Assignment1", "Assignment2", "Assignment3", "Assignment4"]

df["Average"] = df[assignments].mean(axis=1)
df["Median"] = df[assignments].median(axis=1)
df["Maximum"] = df[assignments].max(axis=1)
df["Minimum"] = df[assignments].min(axis=1)

print("Student Statistics:")
df
```

```python
# Calculate assignment statistics

print("Assignment Statistics\n")

for assignment in assignments:
    print(assignment)
    print("Average :", df[assignment].mean())
    print("Median  :", df[assignment].median())
    print("Maximum :", df[assignment].max())
    print("Minimum :", df[assignment].min())
    print("----------------------------")
```

```python
# Sort students by average score

sorted_df = df.sort_values(by="Average", ascending=False)

print("Students Sorted by Average Score:")
sorted_df
```

```python
# Identify the top performer

top_student = sorted_df.iloc[0]

print("Top Performer")
print("Name :", top_student["Name"])
print("Average Score :", round(top_student["Average"], 2))
```

```python
# NumPy operations

scores = df[assignments].to_numpy()

print("Overall Statistics")

print("Overall Average :", np.mean(scores))
print("Overall Highest :", np.max(scores))
print("Overall Lowest :", np.min(scores))
print("Standard Deviation :", round(np.std(scores), 2))
```
